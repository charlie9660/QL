```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import itertools
import mpl_toolkits.mplot3d

#------------------------------------------------------------------------------
# DEFINITION
#------------------------------------------------------------------------------
class QLearning():
    
    
    def __init__(self,data,mc,alpha = 0.1,gamma = 0.9 ,epsilon = 0.1,states = 5,short_w =5,long_w=20,vote = 50):
        np.set_printoptions(threshold=np.nan)
        self.data = data

        self.ret = self.data / self.data.shift(1) - 1
        self.ret = self.ret.fillna(0)
       
        self.window_short = short_w
        self.window_long = long_w
        self.samples = len(self.data)
        self.states = states
        self.actions = 3 #long, flat, short
        self.epsilon = epsilon
        self.gamma = gamma #discount factor
        self.alpha = alpha
        self.mc = mc
        self.vote = vote

        self.q = np.zeros((self.states, self.states, self.actions))
        self.rewards = np.zeros((self.states, self.states, self.actions))
        self.count = np.zeros((self.states, self.states, self.actions), dtype = np.int16)
        self.isVisited = np.zeros((self.states, self.states, self.actions), dtype = np.bool)
        self.actionMap = np.zeros((self.mc,self.samples))
        self.momentum = np.zeros(self.samples)
   
    def init(self):
        self.count = np.zeros((self.states, self.states, self.actions), dtype = np.int16)
        self.isVisited = np.zeros((self.states, self.states, self.actions), dtype = np.bool)
   
    def currentState(self, signal):
        signal = float(signal)
        sep = np.linspace(-1, 1, self.states-1)
        return sum(sep < signal)
        
    def selectAction(self, state_short, state_long):
        if (self.q[state_short, state_long, :]==0).sum() == self.actions:
            #if all action-values are 0
            return np.random.randint(0, self.actions)
        else:
            #Epsilon-Greedy
            if np.random.random(1) < self.epsilon:
                return np.random.randint(0, self.actions)
            else:
                return np.argmax(self.q[state_short, state_long, :])

    def actionToPosition(self, action):
        if action == 0:
            return -1
        elif action == 1:
            return 0
        elif action == 2:
            return 1

    def updateRewards(self, reward, state_short, state_long, action):
        self.isVisited[state_short, state_long, action] = True
        self.rewards = self.rewards + reward * ( self.count)
        self.count = self.count + self.isVisited

    def updateQ(self, itr):
        self.q = (self.q * itr + self.rewards) / (itr + 1)
        #self.q = self.q
        
    def updateQI(self,old_short,old_long,i):
        
        sub_short = self.momentum[i - self.window_short + 1  : i]
        sub_long = self.momentum[i - self.window_long + 1 : i ]
               
        state_short = self.currentState( np.mean(sub_short) / np.std(sub_short) * np.sqrt(252) )
        state_long = self.currentState( np.mean(sub_long) / np.std(sub_long) * np.sqrt(252) )
        #print((old_short,state_short))
        #self.q[old_short, old_long,:] = self.q[old_short, old_long,:] + self.alpha * (self.rewards[old_short, old_long,:] + self.gamma ** max(self.q[state_short, state_long,:]) - self.q[old_short, old_long,:])
        self.q = self.q + self.alpha * (self.rewards + self.gamma ** max(self.q[state_short, state_long,:]) - self.q)

    def epoch(self,index):
        for i in range(self.samples - 1):
            if i <= self.window_long - 1:
                self.momentum[i] = self.ret.ix[i]
            else:
                sub_short = self.momentum[i - self.window_short : i - 1]
                sub_long = self.momentum[i - self.window_long : i - 1]
               
                #state = annualized Sharpe ratio
                state_short = self.currentState( np.mean(sub_short) / np.std(sub_short) * np.sqrt(252) )
                state_long = self.currentState( np.mean(sub_long) / np.std(sub_long) * np.sqrt(252) )

                action = self.selectAction(state_short, state_long)
                self.actionMap[index,i] = action
                reward = self.ret.ix[i + 1] * self.actionToPosition(action)
                self.updateRewards(reward, state_short, state_long, action)
                self.updateQI(state_short,state_long,i)
                self.momentum[i] = reward
            

    def Agent(self):
        best =  np.zeros(self.samples)
        best_q = np.zeros((self.states, self.states, self.actions))
        returns =  np.zeros(self.mc)
        best_index = 0
        comp = 0
        for i in range(self.mc):
            self.init()
            self.epoch(i)
            if(sum(self.momentum) > sum(best) and i > (self.mc - self.vote)):
                best = self.momentum.copy()
                best_q = self.q.copy()
                best_index = i
            #print("episode",i,"done.Best cumulative return is",sum(best))
            #print("episode",i,"done.Cumulative return is",sum(self.momentum))
            if(i>0):
                comp = returns[i-1]
            returns[i] = (comp* i + sum(self.momentum))/ (i+1) 
            print("epoch",i,"done.Average Cumulative return is",returns[i])
            self.updateQ(i)
          
      
        #actions = pd.DataFrame(self.actionMap[-self.vote:,:]).mode().iloc[0]   
        #actions = map(lambda x:self.actionToPosition(x),actions)
       
        best_actions = self.actionMap[best_index,:]
        best_actions = map(lambda x:self.actionToPosition(x),best_actions)
        #voted = actions *self.ret.shift(-1)[:len(actions)]
        plt.plot(100 * (1 + self.momentum).cumprod(), label="RL-momentum-Last")
        #plt.plot(100 * (1 + voted).cumprod(), label="RL-momentum-Vote")
        plt.plot(100 * (1 + best).cumprod(), label="RL-momentum-Best")
        plt.plot(100 * (1 + self.ret).cumprod(), label="long-only")
        plt.plot(100 * (1 - self.ret).cumprod(), label="short-only")
        plt.legend(loc="best")
        plt.show()
        
        plt.plot(returns)
        plt.plot()
        
        self.q = best_q.copy()
       
        #plot Q-value matrix
        x = np.linspace(0,5,self.states)
        y = np.linspace(0,5,self.states)
        x,y = np.meshgrid(x, y)
       
        for i in range(self.actions):
            if i == 0:
                position = "short"
            elif i == 1:
                position = "flat"
            elif i == 2:
                position =  "long"

            fig = plt.figure()
            ax = fig.gca(projection='3d')
           
            ax.set_xlabel("state_short")
            ax.set_ylabel("state_long")
            ax.set_zlabel("Q-value")
            ax.set_title("Q-value for " + position + " position")
            #ax.view_init(90, 90)
            urf = ax.plot_surface(x, y, self.q[:, :, i], rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)

            plt.show()
        
        return(self.q)
    
    def Predict(self,q,df):

        momentum = np.zeros(len(df))
        ret = df / df.shift(1) - 1
        for i in range(len(df) - 1):
            if i <= self.window_long - 1:
                momentum[i] = ret.ix[i+1]
            else:
                sub_short = momentum[i - self.window_short : i - 1]
                sub_long =  momentum[i - self.window_long : i - 1]
               
                #state = annualized Sharpe ratio
                state_short = self.currentState( np.mean(sub_short) / np.std(sub_short) * np.sqrt(252) )
                state_long = self.currentState( np.mean(sub_long) / np.std(sub_long) * np.sqrt(252) )
                self.q = q
                action = self.selectAction(state_short, state_long)
                reward = ret[i] * self.actionToPosition(action)
                #print((ret.ix[i],self.actionToPosition(action)))
                momentum[i] = reward
        return momentum
        
        
    
```

```python
#Training Q network
data = pd.read_csv("data.csv",na_values =".")["DEXUSEU"].fillna(method='ffill')

train_data  = data[100:200].reset_index(drop=True)
m = QLearning(train_data,2000,short_w=5,long_w=10,gamma = 0.8, alpha=0.7,vote=20)
q = np.zeros((5, 5,3))
for i in range(5):
    q = (m.Agent() + q) / (i+1)
    print(q)
#actions = pd.DataFrame(best).mode().iloc[0]   
```

```python
test_data = data.ix[101:120]
test_data= test_data.reset_index(drop=True)
momentum = m.Predict(q,test_data)
print(momentum)
momentum[np.isnan(momentum)] = 0
ret = test_data / test_data.shift(1) - 1
ret = ret.shift(-1)
plt.plot( (1 + momentum).cumprod(), label="QL")
plt.plot( (1 + ret).cumprod(), label="long-only")
plt.plot( (1 - ret).cumprod(), label="short-only")
plt.legend(loc="best")
plt.show()
```
